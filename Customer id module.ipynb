```python
create Table python_db_01.cust_py
(
	cust_id BIGINT NOT NULL,
	st_dt DATETIME NOT NULL,
	first_name Varchar(100) NOT NULL,
    last_name Varchar(100) NOT NULL,
    dob CHAR(11) NULL,
    g Char(1),
	e_dt DATETIME,
    image_path varchar(100),
    PRIMARY KEY (cust_id, st_dt),
	INDEX (cust_id),
    INDEX (first_name, last_name)
);

create Table python_db_01.email
(
	cust_id BIGINT NOT NULL,
	st_dt DATETIME NOT NULL,
	e_dt DATETIME,
    eml varchar(200),
    PRIMARY KEY (cust_id, st_dt),
	INDEX (cust_id)
);

create Table python_db_01.phone
(
	cust_id BIGINT NOT NULL,
	st_dt DATETIME NOT NULL,
	e_dt DATETIME,
    phone varchar(10),
    PRIMARY KEY (cust_id, st_dt),
	INDEX (cust_id)
);

create Table python_db_01.tg
(
	cust_id BIGINT NOT NULL,
	st_dt DATETIME NOT NULL,
	e_dt DATETIME,
    tg_chat_id varchar(10),
    PRIMARY KEY (cust_id, st_dt),
	INDEX (cust_id)
);



create Table python_db_01.address
(
	cust_id BIGINT NOT NULL,
	st_dt DATETIME NOT NULL,
	e_dt DATETIME,
    al1 varchar(200),
    al2 varchar(200),
    al3 varchar(200),
    al4 varchar(200),    
    PRIMARY KEY (cust_id, st_dt),
	INDEX (cust_id)
);


create Table python_db_01.error_log
(
	err_id BIGINT NOT NULL AUTO_INCREMENT,
	er_dt DATETIME NOT NULL,
    er_msg VARCHAR(200),
    er_func VARCHAR(100)
);

```
